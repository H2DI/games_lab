```python
M = 5
K = 5
values = np.zeros((M, 1+(K-1)*M))
print(1+(K-1)*M)

def f(x):
    return np.exp(- np.square((K-1)*x - 1))

for i in range(M):
    if i==0:
        values[i, :] = 1 - f(np.linspace(0, 1, 1+(K-1)*M))
    else: 
        values[i, :] = f(np.linspace(0, 1, 1+(K-1)*M))
plt.plot(values[0])    
plt.plot(values[1])    

game = PartyGame(M, K, values)
#print(values)
```

```python
plays = ()
for _ in range(M):
    play = np.random.rand(K)
    play = play / np.sum(play)
    plays += (play,)

print(game.compute_exp_payoffs(0, plays))
print(game.compute_exp_payoffs(1, plays))
print(game.compute_exp_payoffs(2, plays))
```

```python
agent_list = [ BMAlg(K, label="Player {}".format(i)) for i in range(M)]
T = 30000
instance = MultiAgent(game, agent_list)
instance.play_T_times(T)
```

```python
for agent in agent_list:
    plt.title("Internal regret")
    plt.plot(agent.int_regret, label=agent.label)
    plt.xscale("log")
    plt.legend()
    plt.show()
```

```python
for agent in agent_list:
    plt.title("External regret")
    plt.plot(agent.ext_regret, label=agent.label)
    plt.xscale("log")
    plt.legend()
    plt.show()
```

```python
for agent in agent_list:
    plt.title("Rewards")
    t_slice = range(T)#range(min(agent.times_war)-100, max(agent.times_war) + 100)
    plt.plot(t_slice, np.array(agent.received_rewards)[t_slice], label=agent.label)
    plt.xscale('log')
    plt.legend()
    plt.show()
```

```python
guest_numbers = np.arange(K)
for agent in agent_list:
    plt.title("Expected Guest Numbers (max is {})".format(K))
    t_slice = range(T)#range(min(agent.times_war)-100, max(agent.times_war) + 100)
    plt.plot(t_slice, [np.dot(np.array(agent.play_history[t]), guest_numbers) for t in t_slice], label=agent.label)
    plt.xscale('log')
    plt.legend()
plt.show()
for agent in agent_list:
    plt.title("Entropy of Guest Numbers (max is {})".format(K))
    t_slice = range(T)#range(min(agent.times_war)-100, max(agent.times_war) + 100)
    plt.plot(t_slice, [entropy(agent.play_history[t]) for t in t_slice], label=agent.label)
    plt.xscale('log')
    plt.legend()
plt.show()
```
