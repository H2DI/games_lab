```python
M = 2
K = 4
values = np.zeros((M, 1+(K-1)*M))

def aim(x, a, b):
    if a <= x <= b:
        return 1
    else:
        return 0
                   
xs = np.linspace(0, 1, 1+(K-1)*M)             
for i in range(M):
    if i==0:
        values[i, :] = [ aim(x, 0.2, 0.6) for x in xs]
    elif i==1: 
        values[i, :] = [1 - aim(x, 0.2, 0.6) for x in xs]

plt.plot(values[0], label='Player 0')
plt.plot(values[1], label='Player 1')    

plt.legend()

game = PartyGame(M, K, values)
```

# Compare all algorithms

```python
alg_list = [OptimisticAdaHedge,
            OptimisticBMAdaHedge,
            OptimisticHedge,
            OptimisticHedge_a,
            Hedge,
            Hedge_a, 
            BMAlg,
            BMAlg_a,
            OptimisticBMAlg, 
            OptimisticBMAlg_a]

alg_names = ["Optimistic AdaHedge",
             "Optimistic BM AdaHedge"
             "Optimistic Hedge", 
             "Optimistic Hedge Anytime",
             "Hedge", 
             "Hedge Anytime", 
             "BM",
             "BM anytime",
             "Optimistic BM", 
             "Optimistic BM anytime"]

T = 10000

for i, alg in enumerate(alg_list):
    agent_list = [ alg(K, horizon=T, label="Player {}".format(i)) for i in range(M) ]
    
    instance = MultiAgent(game, agent_list)
    instance.initialize(init_plays)
    instance.play_T_times(T)
    
    guest_numbers = np.arange(K)
    t_slice = range(T)
    total = np.zeros(T)
    for agent in agent_list:
        plt.title("Expected Guest Numbers for {} (max is {})".format(alg_names[i], 2*K-1))
        average_guest_number = np.array([np.dot(np.array(agent.play_history[t]), guest_numbers) for t in t_slice])
        variance = np.array([np.dot(np.array(agent.play_history[t]), np.square(guest_numbers - average_guest_number[t])) for t in t_slice])
        total += average_guest_number
        plt.plot(t_slice, average_guest_number, label=agent.label)
        plt.fill_between(t_slice, average_guest_number - 0.1*variance, average_guest_number + 0.1*variance, alpha=0.2)
        plt.xscale('log')
    plt.plot(t_slice, total, label="Sum") 
    plt.legend()
    plt.show()


#for agent in agent_list:
#    plt.title("Entropy of Guest Numbers (max is {})".format(np.log(K)))
#    plt.plot(t_slice, [entropy(agent.play_history[t]) for t in t_slice], label=agent.label)
#    plt.xscale('log')
#    plt.legend()
#plt.show()
```

```python
M = 3
K = 5
values = np.random.rand(M*(1+(K-1)*M)).reshape((M, 1+(K-1)*M))

def aim(x, a, b):
    if a <= x <= b:
        return 1
    else:
        return 0
                   
xs = np.linspace(0, 1, 1+(K-1)*M)             

plt.plot(values[0], label='Player 0')
plt.plot(values[1], label='Player 1')    
plt.plot(values[2], label='Player 2')    


plt.legend()

game = PartyGame(M, K, values)
```

```python
agent_list = [ OptimisticAdaHedge(K, label="Player {}".format(i)) for i in range(M) ]

T = 100000

instance = MultiAgent(game, agent_list)
instance.play_T_times(T)
```

```python
guest_numbers = np.arange(K)
t_slice = range(T)
total = np.zeros(T)
for agent in agent_list:
    plt.title("Expected Guest Numbers for {} (max is {})".format(agent.label, 2*K-1))
    average_guest_number = np.array([np.dot(np.array(agent.play_history[t]), guest_numbers) for t in t_slice])
    variance = np.array([np.dot(np.array(agent.play_history[t]), np.square(guest_numbers - average_guest_number[t])) for t in t_slice])
    total += average_guest_number
    plt.plot(t_slice, average_guest_number, label=agent.label)
    plt.fill_between(t_slice, average_guest_number - 1*variance, average_guest_number + 1*variance, alpha=0.2)
#plt.xscale('log')
plt.plot(t_slice, total, label="Sum") 
plt.legend()
plt.show()
```

```python
max_internal_regret = 0
for agent in agent_list:
    plt.title("Internal regret")
    plt.plot(agent.int_regret, label=agent.label)
    max_internal_regret = max(max_internal_regret, np.max(agent.int_regret))
    plt.xscale("log")
    plt.legend()
plt.show()

print(max_internal_regret)
```
