```python
def f(K, N=10):
    T = 2*K
    maximums = []
    argmaxs = []
    last_vals = []
    for _ in range(N):
        #K, M = 100, 2
        shape = 2 * [K,] + [2,]
        #game = RandomGame(shape, zerosum=True, integer=True)#FindTheColumn(shape)
        game = GuessTheNumber(K)

        
        agent_list = [OptimisticAdaHedge(K, label="") for _ in range(2)]
        init_plays = [ normalize(np.random.rand(K)) for _ in range(M) ]
        instance = MultiAgent(game, agent_list)
        instance.initialize(init_plays)
        instance.play_T_times(T-1)
        
        reg_sum = np.array(agent_list[0].ext_regret) + np.array(agent_list[1].ext_regret)
        
        maximums.append(np.max(reg_sum))
        argmaxs.append(np.argmax(reg_sum))
        last_vals.append(reg_sum[-1])
    return maximums, argmaxs, last_vals
        
Ks = [2, 3, 5, 10, 15, 20, 30, 50, 100]#, 150, 200, 250]
N = 20

m_mean, a_mean, l_mean = [], [], []
m_var, a_var, l_var = [], [], []

m_max = []


for K in Ks:
    t = time.time()
    m, a, l = f(K, N)
    print(' K = {} took {}'.format(K, time.time() - t))
    m_mean.append(np.mean(m))
    m_max.append(np.max(m))
    a_mean.append(np.mean(a))
    l_mean.append(np.mean(l))
    m_var.append(np.var(m))
    a_var.append(np.var(a))
    l_var.append(np.var(l))
```

## Plotting the cumulative Mix gaps:
i.e. 
$$
\sum_{t=1}^T\bigg( \langle p_t,  \ell_t - \ell_{t-1} \rangle  - \frac{-1}{\eta_t} \sum_{i=1}^K p_{i, t}e^{-\eta_t (\ell_{i,t} - \ell_{i, t-1})} \bigg)
$$

## Plotting the square root of cumulative variances
i.e.
$$
\sqrt{\sum_{t=1}^T \sum_{i = 1}^K p_{t, i} \big( \ell_{t, i} - \ell_{t-1, i}  - \langle p_t, \ell_{t} - \ell_{t-1} \rangle  \big)^2}
 = 
 \sqrt{\sum_{t=1}^T  \mathrm{Var}_{I_t \sim p_t}  \big( \ell_{t, I_t}- \ell_{t-1,I_t} \big)}
$$

```python
for agent in agent_list:
    plt.title("(Optimistic) Cumulative variances")
    t_slice = range(T)
    plt.plot(np.square(np.cumsum(agent.diffs_var)), label=agent.label)
    plt.legend()
plt.xscale('log')
plt.show()
```

```python
import numpy as np
import math
import matplotlib.pyplot as plt
import scipy.linalg
import warnings

def fixed_point(Q):
        """
            Computes the probability fixed point of a stochastic matrix
        """
        K = len(Q)
        X = np.vstack((np.eye(K) - Q, np.ones(K)))
        b = np.zeros(K + 1)
        b[K] = 1

        v = np.matmul(np.transpose(X), X)
        btilde = np.matmul(np.transpose(X), b)

        with warnings.catch_warnings():
            warnings.filterwarnings("error")
            try:
                p = scipy.linalg.solve(v, btilde)
                # if any(p < 0) or any(p > 1) or not (np.isclose(np.sum(p), 1)):
                #     print("Standard solve: ")
                #     print("Q")
                #     print(Q)
                #     print("p")
                #     print(p)
                return p
            except (Warning, np.linalg.LinAlgError) as e:
                # self.times_war.append(self.t)
                return np.sum(np.power(Q, 10), axis=1) / K
            
def kullback(p, q):
    """Kullback-Leibler divergence for Bernoulli distributions."""
    pprime, qprime = np.copy(p), np.copy(q)
    eps = 1e-15

    K = len(p)
    for i in range(K):
        if q[i] < eps:
            if p[i] < eps:
                qprime[i] = 1
                pprime[i] = 1
            else:
                return 1e15
    return np.dot(p, np.log(pprime / qprime))
```

```python
def random_p(K):
    r = np.random.rand(K)
    return r / np.sum(r)

def random_Q(K):
    l = [random_p(K) for _ in range(K)]
    Q = None
    for p in l:
        if Q is None:
            Q = np.transpose([p])
        else:
            Q = np.hstack((Q, np.transpose([p])))
    return Q

K = 10
for _ in range(100):
    Q = random_Q(K)
    print(np.linalg.det( np.matmul(Q, np.transpose(Q)) + np.ones((K, K))))
    print(np.linalg.det( np.matmul(np.transpose(Q), Q) + np.ones((K, K))))
```

```python
def gen_KL(Q, Qprime):
    r = 0
    for j in range(K):
        r += kullback(Q[:, j], Qprime[:, j])
    return r


def mix_gap(p, A, M, Mprime, eta=1):
    q, qprime = fixed_point(M), fixed_point(Mprime)
    ell = np.dot(A, q - qprime)
    v_exp = np.exp(- eta * ell)
    return np.dot(p, ell) + 1 / eta * np.log(
        np.dot(p, v_exp)
    )
```

```python
def Fij(K, i=-1 ,j=-1):
    M = np.zeros((K, K))
    if i == -1:
        i = np.random.randint(K)
    if j == -1:
        j = np.random.randint(K-1)
        if j == i:
            j += 1
    for k in range(K):
        if k == i:
            M[j, i] = 1
        else:
            M[k, k] = 1
    return M

def AverageFij(K, p = None):
    if p is None:
        p = (np.ones((K, K)) - np.eye(K)) / (K*(K-1))

    M = np.zeros((K, K))
    for i in range(K):
        for j in range(K):
            M += p[i, j] * Fij(K, i, j)
    return M

def CircFij(K, shift=1):
    p = np.eye(K) / K
    p = np.roll(p, shift, axis=1)
    return AverageFij(K, p=p)

def RandomFij(K): # Generates a random convex combination of the Fij's
    p = np.random.rand(K, K)
    for i in range(K):
        p[i, i] = 0
    p = p / np.sum(p)
    return AverageFij(K, p = p)
```

```python
K = 3

M = Fij(K, 0, 1)
H = RandomFij(K) #AverageFij(K)

u = np.ones(K) / K
A = np.random.rand(K*K).reshape((K, K))#np.eye(K)

r1, r2 = [], []
alphas = np.linspace(0.1, .9, 30)
for alpha in alphas:
    Mprime= (1-alpha)*M + alpha*H
    r1.append(gen_KL(M, Mprime))
    r2.append(mix_gap(u, A, M, Mprime, eta=1))

plt.plot(alphas, r1, label='KL')
plt.plot(alphas, r2, label='Mix gaps')
plt.title("K="+str(K))
plt.legend()
plt.show()
```

```python
def randommat(K):
    P = np.eye(K)
    b = np.arange(K)
    np.random.shuffle(b)
    P = P[b, :]

    A = np.ones((K, K))
    for i in range(K):
        for j in range(K):
            if i > j:
                A[i, j] = -1
            elif i == j:
                A[i, j] = 0
    mat = np.dot(np.transpose(P), np.dot(A, P))
    return mat, b
```
