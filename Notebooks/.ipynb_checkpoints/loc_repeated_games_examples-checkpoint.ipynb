Declare the game we are going to play, and the number of actions.

## Plotting the cumulative Mix gaps:
i.e. 
$$
\sum_{t=1}^T\bigg( \langle p_t,  \ell_t - \ell_{t-1} \rangle  - \frac{-1}{\eta_t} \sum_{i=1}^K p_{i, t}e^{-\eta_t (\ell_{i,t} - \ell_{i, t-1})} \bigg)
$$

```python
t_slice = range(T-1)
for j, agent in enumerate(agent_list):
    plt.title("(Optimistic) Cumulative Mix Gaps")
    plt.plot(t_slice, np.cumsum(agent.mix_gaps)[t_slice], label=agent.label)
    plt.legend()
plt.xscale('log')
plt.show()
```

## Plotting the square root of cumulative variances
i.e.
$$
\sqrt{\sum_{t=1}^T \sum_{i = 1}^K p_{t, i} \big( \ell_{t, i} - \ell_{t-1, i}  - \langle p_t, \ell_{t} - \ell_{t-1} \rangle  \big)^2}
 = 
 \sqrt{\sum_{t=1}^T  \mathrm{Var}_{I_t \sim p_t}  \big( \ell_{t, I_t} - \ell_{t-1,I_t} \big)}
$$

```python
for agent in agent_list:
    plt.title("(Optimistic) Cumulative variances")
    t_slice = range(T)
    plt.plot(np.square(np.cumsum(agent.diffs_var)), label=agent.label)
    plt.legend()
plt.xscale('log')
plt.show()
```

```python
a1, a2 = agent_list[0], agent_list[1]

l1 = a1.play_history 
l2 = a2.play_history

t = len(l2)

a = []
b = []
for s in range(1,t): 
    a.append(np.dot(l1[s], np.square( l2[s] - l2[s-1] - np.dot(l1[s], l2[s] - l2[s-1]))))
    b.append(np.dot(l2[s], np.square( l1[s] - l1[s-1] - np.dot(l2[s], l1[s] - l1[s-1]))))

    
    
plt.plot(np.cumsum(a))
plt.plot(np.cumsum(b))

plt.xscale('log')
```

## Plotting the square root of cumulative second moment
i.e.
$$
\sqrt{\sum_{t=1}^T \sum_{i = 1}^K p_{t, i} \big( \ell_{t, i} - \ell_{t-1, i}   \big)^2}
 \geqslant
 \sqrt{\sum_{t=1}^T \mathrm{Var}_{I_t \sim p_t}  \big( \ell_{t, I_t} - \ell_{t-1,I_t} \big)}
$$

In general, this seems to go to infinity (although quite slowly). Thus the upper bound looks loose.

```python
for agent in agent_list:
    #agent = agent_list[-1]
    plt.title("(Optimistic) Cumulative second moment")
    t_slice = range(T)
    plt.plot(np.sqrt(np.cumsum(agent.diffs_sec_moment)), label=agent.label)
    plt.legend()
plt.xscale('log')
plt.show()
```

```python
plt.title("(Optimistic) Cumulative second moment")
for agent in agent_list:
    agent.plot_regret_bound()
    plt.legend()
plt.xscale('log')
plt.show()
```
